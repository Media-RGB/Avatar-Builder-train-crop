```python
def create_flow_shape(facial, facial_df):
    facial_df = df_hairstyle[facial_df]
    files = df_hairstyle['hair_path']
    
    img_nb = 0
    for attribute in facial_df:
        img_cv = cv2.imread(files[img_nb])
        img_r = cv2.resize(img_cv, (256, 256), interpolation = cv2.INTER_AREA)
        
        ATTRIBUTE_DIR = os.path.join(CROPPED_FF_DIR, f'{database}//t_{facial}//{attribute}')
        if not os.path.exists(ATTRIBUTE_DIR):
            os.makedirs(ATTRIBUTE_DIR)
        img_file = os.path.join(ATTRIBUTE_DIR, f'{facial}_{img_nb:04d}.png')
        cv2.imwrite(img_file, img_r)
        
        img_nb += 1
```

```python
def create_flow(facial, facial_df):
    facial_df = csv[facial_df][:nb_imgs]
    files_obj = Files(facial, parent_dir=CROPPED_FF_DIR)
    
    img_nb = 0
    for attribute in facial_df:
        img = cv2.imread(files_obj.files[img_nb])
        img_obj = Image(img)
        
        img_obj.resize_img(size=(SIZE,SIZE))
        img_obj.gray_color()
        ATTRIBUTE_DIR = os.path.join(CROPPED_FF_DIR, f'{database}//{facial}_f0//{attribute}')
        if not os.path.exists(ATTRIBUTE_DIR):
            os.makedirs(ATTRIBUTE_DIR)
        img_file = os.path.join(ATTRIBUTE_DIR, f'{facial}_{img_nb:04d}.png')
        cv2.imwrite(img_file, img_obj.img)
        
        img_nb += 1
```

```python
def create_subflow(facial, facial_df, facial_df_sub):
    
    facial_df = csv[facial_df]
    files = csv['hairtop_path']
    facial_df_sub = csv[facial_df_sub]
    
    img_nb = 0
    for attribute, sub_attribute in zip(facial_df, facial_df_sub):
        img = cv2.imread(files[img_nb])
        img_obj = Image(img)
        
#         img_obj.resize_img((128, 128))
        
        ATTRIBUTE_DIR = os.path.join(CROPPED_FF_DIR, f'{database}//{facial}//{attribute}_final2//{sub_attribute}')
        if not os.path.exists(ATTRIBUTE_DIR):
            os.makedirs(ATTRIBUTE_DIR)
        img_file = os.path.join(ATTRIBUTE_DIR, f'{facial}_{img_nb:04d}.png')
        cv2.imwrite(img_file, img_obj.img)
        
        img_nb += 1
```
