```python
dirlist = os.listdir(train_path)
```

```python
dirlist
```

```python
aug_dir = os.path.join(CROPPED_FF_DIR, f'{database}//{facial}_aug2')
if not os.path.exists(aug_dir):
    os.makedirs(aug_dir)
```

```python
for atri in dirlist:
    atri_aug_dir = os.path.join(aug_dir, f'{atri}')
    atri_dir = os.path.join(CROPPED_FF_DIR, f'{database}//{facial}//{atri}')
    if not os.path.exists(atri_aug_dir):
        os.makedirs(atri_aug_dir)
    atri_files = os.listdir(atri_dir)
    nb_batch = math.ceil(BS/len(atri_files))
    for atri_file in atri_files:
        atri_img = os.path.join(atri_dir, f'{atri_file}')
        img = load_img(atri_img)  
        x = img_to_array(img) 
        # Reshape the input image 
        x = x.reshape((1, ) + x.shape)  
        i = 0
        print(nb_batch)
        # generate 10 new augmented images 
        for batch in datagen.flow(x, batch_size = 1, 
                          save_to_dir =atri_aug_dir,  
                          save_prefix =f'{atri}_{i:04d}', save_format ='png'):
            i += 1
            if i > nb_batch: 
                break

        
        
                                         
                                         
                                         
```
